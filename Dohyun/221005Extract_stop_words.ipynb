```python
import pandas as pd
from soynlp.noun import LRNounExtractor_v2
from soynlp.tokenizer import LTokenizer
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/LIFLOW/data/all_df_with_tokens_according_to_topic.csv')
print(len(df))
df.head()
```

```python
sentences = list(df['sentence'])
noun_extractor = LRNounExtractor_v2(verbose=True)
nouns = noun_extractor.train_extract(sentences)

noun_score = dict([])
for i in nouns :
  noun_score[i] = nouns[i].score
```

```python
tokenizer = LTokenizer(scores = noun_score)
tokens = [tokenizer.tokenize(s) for s in sentences]
```

```python
# Should use CountVectorizer with sklearn, but we already tokenize with soynlp
# So, simply count tokens. 
voca = dict([])
for i in range(len(tokens)) :
  for j in range(len(tokens[i])) :
    if tokens[i][j] not in voca :
      voca[tokens[i][j]] = 1
    else :
      voca[tokens[i][j]] += 1
```

```python
len(voca)
```

```python
def extract_stop_words(voca, start, end) : 
  res = sorted(voca.items(), key = lambda x : x[1], reverse = True)
  return res[start:end]
```

```python
extract_stop_words(voca, 200, 300)
```
