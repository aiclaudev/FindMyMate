```python
import pandas as pd
from konlpy.tag import *
from soynlp.noun import LRNounExtractor_v2
from soynlp.tokenizer import LTokenizer
```

```python
df = pd.read_csv('C:/Users/82102/Desktop/Project/LIFLOW/Data/AI_chatBot.csv')
```

```python
sentences = list(df['sentence'])
sentences[:10]
```

```python
# NO : 외부 잡음
# FP : 간투어 (음, 아, 그)
# SP : 발음이 정확히 들리지 않는 경우 (세시? 네시?)
# SN : 화자잡음
```

```python
NOs = []
FPs = []
SPs = []
SNs = []

for i in sentences :
    if '(NO' in i :
        NOs.append(i+'\n')
    if '(FP' in i :
        FPs.append(i+'\n')
    if '(SP' in i :
        SPs.append(i+'\n')
    if '(SN' in i :
        SNs.append(i+'\n')
```

```python
with open('NOs', 'w', encoding='UTF8') as f :
    f.writelines(NOs)
with open('FPs', 'w', encoding='UTF8') as f :
    f.writelines(FPs)
with open('SPs', 'w', encoding='UTF8') as f :
    f.writelines(SPs)
with open('SNs', 'w', encoding='UTF8') as f :
    f.writelines(SNs)
```

## konlpy

```python
okt = Okt()
komoran = Komoran()
kkma = Kkma()
```

```python
okt_tokens = [okt.morphs(sent) for sent in sentences[:200]]
komoran_tokens = [komoran.morphs(sent) for sent in sentences[:200]]
kkma_tokens = [kkma.morphs(sent) for sent in sentences[:200]]
```

```python
print(okt_tokens[:5])
print(komoran_tokens[:5])
print(kkma_tokens[:5])
```

```python
okt_txt = [' '.join(sent) + '\n' for sent in okt_tokens]
komoran_txt = [' '.join(sent) + '\n' for sent in komoran_tokens]
kkma_txt = [' '.join(sent) + '\n' for sent in kkma_tokens]
```

```python
with open('okt_tokenization_result.txt', 'w', encoding='UTF8') as f :
    f.writelines(okt_txt)
with open('komoran_tokenization_result.txt', 'w', encoding='UTF8') as f :
    f.writelines(komoran_txt)
with open('kkma_tokenization_result.txt', 'w', encoding='UTF8') as f :
    f.writelines(kkma_txt)
```

## soynlp

```python
noun_extractor = LRNounExtractor_v2(verbose=True)
nouns = noun_extractor.train_extract(sentences, min_noun_frequency=5)
```

```python
noun_score = {}
for i in nouns :
    noun_score[i] = float(nouns[i].score)
noun_score
```

```python
tokenizer = LTokenizer(scores = noun_score)
soynlp_tokens = [tokenizer.tokenize(sent) for sent in sentences]
```

```python
soynlp_txt = [' '.join(w) + '\n' for w in soynlp_tokens]
```

```python
with open('soynlp_tokenization_result.txt', 'w', encoding='UTF8') as f :
    f.writelines(soynlp_txt)
```
