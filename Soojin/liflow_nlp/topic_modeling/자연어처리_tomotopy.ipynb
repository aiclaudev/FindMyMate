```python
import re

text_origin = files_thema1['sentence']
remain_pattern = re.compile(r'\([^)]*\)')
hangul = re.compile('[^ ㄱ-ㅣ가-힣]+')

text_rm = []

for txt in text_origin:
    ws_replace = re.sub(r'\xa0', r'', txt)
    text_rm.append(remain_pattern.sub(r'', ws_replace))

text = []
for txt in text_rm:
    ws_replace = re.sub(r'\xa0', r'', txt)
    text.append(hangul.sub(r'', ws_replace))

files_thema1['sentence'] = text

files_thema1['sentence']
```

```python
from konlpy.tag import Okt, Kkma, Mecab, Komoran, Hannanum

df_noun1 = []
for i in files_thema1['sentence']:
  a = Komoran().nouns(i)
  p = []
  for j in a:
    if len(j)!=1:
      p.append(j)
  df_noun1.append(p)
print(df_noun1)
```

```python
import tomotopy as tp # 먼저 모듈을 불러와야겠죠
 
# LDAModel을 생성합니다.
# 토픽의 개수(k)는 20개, alpha 파라미터는 0.1, eta 파라미터는 0.01
# 전체 말뭉치에 5회 미만 등장한 단어들은 제거
# 가장 많이 등장하는 최상위 단어 30개 제거
 
model = tp.LDAModel(k=20, alpha=0.1, eta=0.01, min_cf=5, rm_top = 30 )
model.add_doc(df_noun1) # tokenize함수를 이용해 전처리한 결과를 add_doc에 넣습니다.
 
model.train(0)
print('Total docs:', len(model.docs))
print('Total words:', model.num_words)
print('Vocab size:', model.num_vocabs)
 
 
for i in range(200):
    print('Iteration {}\tLL per word: {}'.format(i, model.ll_per_word))
    model.train(1)
 
for i in range(model.k):
    res = model.get_topic_words(i, top_n=10)
    print('Topic #{}'.format(i), end='\t')
    print(', '.join(w for w, p in res))
```
