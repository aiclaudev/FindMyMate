```python
from konlpy.tag import Mecab

mecab = Mecab()
def mecab_tokenize(sent):
  w = []
  a = re.compile('[^ ㄱ-ㅣ가-힣]+')
  words = mecab.pos(sent, join=True)
  word = [a.sub(r'',w) for w in words if ('/NNG' in w or '/NNP' in w)]
  #print(word)
  for i in range(len(word)):
    #print(len(word[i]))
    if len(word[i])!=1:
      w.append(word[i])
  #print(w)
  return w
```
